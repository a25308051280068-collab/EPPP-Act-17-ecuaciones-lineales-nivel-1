Problemas de Simplificación (Términos Semejantes)
Cuando tenemos expresiones como las de los problemas 3 y 5, el objetivo es reducir la expresión sumando o restando términos que tengan la misma letra.
x + 8



```python
# Pedimos los números del problema
numero_suma = int(input("Introduce el número que suma (ej. 8): "))
total = int(input("Introduce el resultado de la igualdad: "))

# Python busca la x restando
x = total - numero_suma
print(f"El valor buscado de x es: {x}")
```

#### Proble 3
Problema 3: $3x + 4x$Planteamiento: Se tienen dos grupos del mismo objeto ($x$).Ecuación/Simplificación: Se suman los coeficientes (números): $3 + 4 = 7$.Resultado: $7x$.

```python
# Pedimos los coeficientes de x
n1 = int(input("Introduce el primer número de x (ej. 3): "))
n2 = int(input("Introduce el segundo número de x (ej. 4): "))
total = int(input("Introduce el resultado total: "))

# Python suma las x y divide
x = total / (n1 + n2)
print(f"El valor buscado de x es: {x}")
```

#### Problema 5
Problema 5: $6a - 2a + a$Planteamiento: Operaciones sucesivas con la variable $a$. (Recuerda que $+a$ es lo mismo que $+1a$).Simplificación: Primero restamos: $6a - 2a = 4a$. Luego sumamos: $4a + 1a = 5a$.Resultado: $5a$.
6a - 2a + a

```python
# Pedimos los números que acompañan a la 'a'
n1 = int(input("Introduce el primer número (ej. 6): "))
n2 = int(input("Introduce el número que resta (ej. 2): "))
total = int(input("Introduce el resultado total: "))

# Despeje: (6 - 2 + 1)a = total
a = total / (n1 - n2 + 1)
print(f"El valor buscado de a es: {a}")
```

#### Problema 7
Problema 7: $4x + 3 = 15$Despeje: $4x = 15 - 3 \rightarrow 4x = 12$.División: $x = 12 / 4$.Solución: $x = 3$.



```python
# Pedimos el múltiplo y la suma
multiplo = int(input("Introduce el número que multiplica a x (ej. 4): "))
suma = int(input("Introduce el número que suma (ej. 3): "))
total = int(input("Introduce el resultado total: "))

# Despeje: x = (total - suma) / multiplo
x = (total - suma) / multiplo
print(f"El valor buscado de x es: {x}")
```
