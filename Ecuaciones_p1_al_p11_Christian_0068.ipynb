#### Proble 3
Problema 3: $3x + 4x$Planteamiento: Se tienen dos grupos del mismo objeto ($x$).Ecuación/Simplificación: Se suman los coeficientes (números): $3 + 4 = 7$.Resultado: $7x$.

```python
# Solicitamos el valor de x
x = int(input("Introduce el valor de x para (3x + 4x): "))

# Al simplificar sabemos que 3x + 4x = 7x
resultado = (3 * x) + (4 * x)

print(f"El resultado de sumar 3({x}) + 4({x}) es: {resultado}")
```

#### Problema 5
Problema 5: $6a - 2a + a$Planteamiento: Operaciones sucesivas con la variable $a$. (Recuerda que $+a$ es lo mismo que $+1a$).Simplificación: Primero restamos: $6a - 2a = 4a$. Luego sumamos: $4a + 1a = 5a$.Resultado: $5a$.
6a - 2a + a

```python
# Solicitamos el valor de a
a = int(input("Introduce el valor de a para (6a - 2a + a): "))

# Realizamos las operaciones en orden
resultado = (6 * a) - (2 * a) + a

print(f"El resultado final de la expresión para a = {a} es: {resultado}")
```

#### Problema 7
Problema 7: $4x + 3 = 15$Despeje: $4x = 15 - 3 \rightarrow 4x = 12$.División: $x = 12 / 4$.Solución: $x = 3$.



```python
# Entrada de datos para x
x = int(input("Introduce el valor de x para (4x + 3): "))

# Primero multiplica por 4 y luego suma 3
resultado = (4 * x) + 3

print(f"El resultado de 4({x}) + 3 es: {resultado}")
```
