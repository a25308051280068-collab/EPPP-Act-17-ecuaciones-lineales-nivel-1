#### Proble 3
Problema 3: $3x + 4x$Planteamiento: Se tienen dos grupos del mismo objeto ($x$).Ecuación/Simplificación: Se suman los coeficientes (números): $3 + 4 = 7$.Resultado: $7x$.

```python
# Problema 3: 3x + 4x
# Simplificación matemática: 7x
x = 2
resultado3 = (3 * x) + (4 * x)
print(f"Resultado Problema 3 (Simplificado es 7x): {resultado3}")
```

#### Proble 5
Problema 5: $6a - 2a + a$Planteamiento: Operaciones sucesivas con la variable $a$. (Recuerda que $+a$ es lo mismo que $+1a$).Simplificación: Primero restamos: $6a - 2a = 4a$. Luego sumamos: $4a + 1a = 5a$.Resultado: $5a$.
6a - 2a + a

```python

# Problema 5: 6a - 2a + a
# Simplificación matemática: 4a + a = 5a
a = 4
resultado5 = (6 * a) - (2 * a) + a
print(f"Resultado Problema 5 (Simplificado es 5a): {resultado5}")
```
